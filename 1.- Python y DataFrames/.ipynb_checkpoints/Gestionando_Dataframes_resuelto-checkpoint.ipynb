## Df .loc e .iloc

### Seleccionar filas y columnas en Pandas mediante posición con iloc

El método iloc se utiliza en los DataFrames para seleccionar los elementos en base a su ubicación. Su sintaxis es data.iloc[, ], donde y son la posición de las filas y columnas que se desean seleccionar en el orden que aparecen en el objeto. Una notación familiar para los usuarios de Matlab. En un DataFrame cada una de las filas tiene un número que va desde 0 hasta el total de filas menos uno. Siendo iloc el método que permite seleccionar los datos en base a estos números. Lo mismo que aplica a las columnas.

En iloc hay dos argumentos filas y columnas. En caso de que se omita el segundo, se seleccionarán todas las columnas de la fila. Así para seleccionar diferentes filas se puede utilizar.

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.iloc.html#pandas.DataFrame.iloc

### Seleccionar filas y columnas en Pandas en base a etiquetas con loc

El método loc permite seleccionar filas o columnas en base a una condición.


https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.loc.html

Rellena los Nulos en Income según la media de GENDER

```python
# Rellenar con la media
df['Income'].fillna(df['Income'].mean())
#Rellenar con valor fijo
df['Income'].fillna(0)
```

```python
df['Income'].mean()
```

```python
media_male = df[['GENDER_clean','Income']][df['GENDER_clean']=='Male'].mean()
media_female = df[['GENDER_clean','Income']][df['GENDER_clean']=='Female'].mean()
df.loc[(df['Income'].isnull())&(df['GENDER_clean']=='Male'), 'Income'] = media_male
df.loc[(df['Income'].isnull())&(df['GENDER_clean']=='Female'), 'Income'] = media_female
```

## INDICES

Número de serie de cada línea de un dataframe. Originalmente, cuando importas o creas una nueva tabla, es una columna que va desde 0 hasta la longitud (número de líneas) del propio dataframe.

En ocasiones nos interesa que una de nuestras columnas sea el índice, para ello existe el método .set_index('columna'), que nos permite cambiar el índice de la tabla.

A través del método .reset_index() volvemos la acción para atrás

```python
df[df.index<5] 
```

```python
df.set_index(['Customer','ST'],inplace=True)
```

```python
df
```

```python
df['Index'] = df.index
df
```
